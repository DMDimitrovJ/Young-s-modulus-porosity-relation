```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics
import pymc3 as pm
%matplotlib inline
```

```python
raw_density =np.array([6.60, 6.73, 6.74,6.75,6.84, 6.89, 6.90, 6.88, 7.01,7.03]).reshape(-1,1)
rel_density = raw_density/7.86
E_modulus =np.array([135.502, 136.828,136.644,137.792,148.436,147.230,147.278,147.120,147.677,158.005]).reshape(-1,1)
rel_modulus = E_modulus/212
```

```python
# Test data
Xtrain=rel_density 
Ytrain=rel_modulus
n=len(Xtrain)
```

```python
def kernel(a, b, l=0.5, sig_var=0.5):
    distance =sklearn.metrics.pairwise_distances(a,b)
    return sig_var * np.exp(-1/(2*l**2)*distance**2)
```

```python
n_t=50
Xtest = np.linspace(0.7, 1, n_t).reshape(-1,1)
```

```python
K_ss = kernel(Xtest, Xtest)
K_ss.shape
```

```python
# Sample 3 sets of standard normals for our test points,
# multiply them by the square root of the covariance matrix
f_prior = np.random.multivariate_normal(mean = np.zeros(n_t), cov =K_ss, size=3).T
# Now let's plot the 3 sampled functions.
plt.plot(Xtest, f_prior)
#plt.axis([-0.05, 0.05])
plt.title('Three samples from the GP prior')
plt.show()
```

```python
noise =0.001
K = kernel(Xtrain, Xtrain)
Kinv=np.linalg.inv(K + noise * np.eye(n))

K_s = kernel(Xtrain, Xtest)

mu =K_s.T@Kinv@Ytrain.squeeze()
cov = K_ss - K_s.T@Kinv@K_s
stdv = np.sqrt(np.diag(cov).squeeze())
```

```python
f_post = np.random.multivariate_normal(mean = mu, cov =cov, size=3).T
```

```python
plt.plot(Xtest, mu, 'r--', lw=2)
plt.plot(Xtest, f_post)
plt.gca().fill_between(Xtest.flat, mu-2*stdv, mu+2*stdv, color ='#dddddd')
plt.scatter(Xtrain, Ytrain)
```

```python
import statsmodels.api as sm
Xtrain_p =1-Xtrain

Xm = (Xtrain_p) #np.column_stack(Xtrain_p**2

X = sm.add_constant(Xm) # adding a constant


res = sm.OLS(Ytrain, X).fit()
predictions = res.predict(X) 
 
print_model = res.summary(alpha = 0.1)
print(print_model)
resid=res.outlier_test()
```

```python
from statsmodels.sandbox.regression.predstd import wls_prediction_std
prstd, iv_l, iv_u = wls_prediction_std(res)

fig, ax = plt.subplots(figsize=(8,6))

ax.plot(Xtrain_p, Ytrain, 'o', label="data")
#ax.plot(Xtrain, y_true, 'b-', label="True")
ax.plot(Xtrain_p, res.fittedvalues, 'r--.', label="OLS")
ax.plot(Xtrain_p, iv_u, 'c--')
ax.plot(Xtrain_p, iv_l, 'c--')
ax.legend(loc='best');
```

```python
Xdata = np.column_stack((Xtrain_p, Xtrain_p**2))
```

```python
with pm.Model() as model:
          
    ke = pm.Normal('ke', mu=0, sd=3, shape =1)
    sigma = pm.HalfCauchy('sigma', beta=10)
    beta = [-ke-1, ke]
    

    Xshared=pm.Data('porosity', Xdata)

        #link function
    mu = 1 +pm.math.dot(Xshared, beta)

    #likelihood
    y = pm.Normal('y', mu =mu, sd= sigma, observed = Ytrain)
```

```python
with model:
    prior = pm.sample_prior_predictive(samples=1000, random_seed=1235)
    ypr = prior.pop("y")
    true_parameters = prior
```

```python
with model:
    trace=pm.sample(5000)
```

```python
with model:
    ppc = pm.sample_posterior_predictive(trace, random_seed=1235, samples=1000)
```

```python
import arviz as az

with model:   
    idata = az.from_pymc3(trace, posterior_predictive=ppc)
```
